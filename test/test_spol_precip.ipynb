```python
# Fix string dtypes before writing to Zarr
import numpy as np

def fix_string_encoding(ds):
    """Convert fixed-length unicode strings to object dtype for Zarr compatibility"""
    ds = ds.copy()  # Create a mutable copy
    for var in ds.variables:
        if ds[var].dtype.kind == 'U':  # Unicode string
            # print(ds)
            # print(var)
            ds[var] = ds[var].astype(str).astype(object)
    return ds

# Apply to all nodes in the datatree
dtree_fixed = dtree.map_over_datasets(fix_string_encoding)
dtree_fixed
```

## Create separate kerchunk references for each group/sweep

```python
# First, find all groups in the netCDF file
import h5py

local_file_path = '/lustre/desc1/scratch/chiaweih/SPol.precip/cfrad.20220525_015443.665_to_20220525_015709.985_SPOL_PrecipRhi2_RHI_datatree.nc'

with h5py.File(local_file_path, 'r') as f:
    print("Available groups:")
    def list_groups(name, obj):
        if isinstance(obj, h5py.Group):
            print(f"  {name}")
    f.visititems(list_groups)
```

```python
# CORRECTED: Filter the full reference to extract only specific group data
from kerchunk.hdf import SingleHdf5ToZarr
import json
import os

local_file_path = '/lustre/desc1/scratch/chiaweih/SPol.precip/cfrad.20220525_015443.665_to_20220525_015709.985_SPOL_PrecipRhi2_RHI_datatree.nc'
output_dir = '/lustre/desc1/scratch/chiaweih/SPol.precip/kerchunk_refs/'
os.makedirs(output_dir, exist_ok=True)

# Create the full reference ONCE (not in a loop)
print("Creating full kerchunk reference...")
with open(local_file_path, "rb") as infile:
    kchunks = SingleHdf5ToZarr(infile, local_file_path, inline_threshold=300, vlen_encode='leave')
    full_kerchunk_dict = kchunks.translate()

# Now filter the reference for each sweep group
import h5py
with h5py.File(local_file_path, 'r') as f:
    sweep_groups = [key for key in f.keys() if key.startswith('sweep_')]
    print(f"\nFound {len(sweep_groups)} sweeps: {sweep_groups}")

for sweep in sweep_groups:
    # Filter refs to only include this sweep's data
    sweep_refs = {}
    sweep_refs['version'] = full_kerchunk_dict.get('version', 1)
    sweep_refs['refs'] = {}
    
    # Copy only the keys that belong to this sweep
    for key, value in full_kerchunk_dict['refs'].items():
        if key.startswith(f'{sweep}/') or key == '.zgroup' or key == '.zattrs':
            sweep_refs['refs'][key] = value
    
    # Save filtered reference
    json_file = f'{output_dir}{sweep}_reference.json'
    with open(json_file, "w") as f:
        json.dump(sweep_refs, f, indent=2)
    
    print(f"Created reference for {sweep}: {json_file}")
    print(f"  - Contains {len(sweep_refs['refs'])} keys")
```

## Solution: Strip the group prefix to access variables directly

```python
# Create a reference with sweep_0 prefix STRIPPED (variables at root level)
from kerchunk.hdf import SingleHdf5ToZarr
import json
import os

local_file_path = '/lustre/desc1/scratch/chiaweih/SPol.precip/cfrad.20220525_015443.665_to_20220525_015709.985_SPOL_PrecipRhi2_RHI_datatree.nc'
output_dir = '/lustre/desc1/scratch/chiaweih/SPol.precip/kerchunk_refs/'

# Load the full reference (if not already loaded)
if 'full_kerchunk_dict' not in locals():
    with open(local_file_path, "rb") as infile:
        kchunks = SingleHdf5ToZarr(infile, local_file_path, inline_threshold=300, vlen_encode='leave')
        full_kerchunk_dict = kchunks.translate()

# Create a reference for sweep_0 with the prefix stripped
sweep = 'sweep_0'
sweep_refs_flat = {}
sweep_refs_flat['version'] = full_kerchunk_dict.get('version', 1)
sweep_refs_flat['refs'] = {}

# Strip the 'sweep_0/' prefix from all keys
for key, value in full_kerchunk_dict['refs'].items():
    if key.startswith(f'{sweep}/'):
        # Remove the 'sweep_0/' prefix
        new_key = key.replace(f'{sweep}/', '', 1)
        sweep_refs_flat['refs'][new_key] = value
    elif key == '.zgroup' or key == '.zattrs':
        sweep_refs_flat['refs'][key] = value

# Save the flattened reference
json_file = f'{output_dir}{sweep}_reference_flat.json'
with open(json_file, "w") as f:
    json.dump(sweep_refs_flat, f, indent=2)

print(f"Created flattened reference: {json_file}")
print(f"  - Contains {len(sweep_refs_flat['refs'])} keys")
print(f"\nSample keys (first 10):")
for i, key in enumerate(list(sweep_refs_flat['refs'].keys())[:10]):
    print(f"  {key}")
```
