```python
import numpy as np
import matplotlib.pyplot as plt
import cmweather

dbz = ds_zarr["sweep_0"].DBZ
rEarth = 6371000.0  # meters

r = np.asarray(dbz.range.values).reshape(1, -1)  # (1, n_range)
e = np.deg2rad(np.asarray(dbz.elevation.values).reshape(-1, 1))  # (n_azimuth, 1)
rhi_alt_km = (np.sqrt(r**2 + rEarth**2 + 2*r*rEarth*np.sin(e)) - rEarth) / 1000.0

dbz = dbz.assign_coords(rhi_alt=(("azimuth", "range"), rhi_alt_km))
range_vals = dbz.range.values / 1000.0  # Convert to km if needed
rhi_alt_vals = dbz.rhi_alt.values  # Already in km


dbz_vals = dbz.values
R, A = np.meshgrid(range_vals, rhi_alt_vals[:,0])

fig, ax = plt.subplots(figsize=(20, 4))
cf = ax.contourf(R, rhi_alt_vals, dbz_vals, 
                  cmap="ChaseSpectral",
                  levels=np.arange(-10, 55, 1))

ax.set_xlabel("Range (km)")
ax.set_ylabel("Altitude (km)")
plt.colorbar(cf, ax=ax, label="Reflectivity (dBZ)")
plt.ylim(0,15)
plt.xlim(0,150)
plt.show()
```

```python
def calculate_rhi_altitude(r,e):
    """Calculate RHI altitude (kilometers) from range (meters) and elevation angle (radians). 
    """
    rEarth = 6371000.0  # meters
    e_deg = np.deg2rad(e)
    return (np.sqrt(r**2 + rEarth**2 + 2*r*rEarth*np.sin(e_deg)) - rEarth) / 1000.0  # output in km
```

```python
# extract sweep_0
ds_sweep0 = ds_zarr["sweep_0"].to_dataset()

# calculate RHI altitude
da_rhi_alt = calculate_rhi_altitude(ds_sweep0['range'], ds_sweep0['elevation']).compute()

# add RHI altitude as a coordinate
ds_sweep0['rhi_alt'] = da_rhi_alt
ds_sweep0 = ds_sweep0.set_coords('rhi_alt')

# convert range to km
ds_sweep0['range'] = ds_sweep0['range'] / 1000.0  # Convert to km
```

```python
# update dataset attributes
ds_sweep0['rhi_alt'].attrs = {}
ds_sweep0['rhi_alt'].attrs['long_name'] = 'RHI Altitude'
ds_sweep0['rhi_alt'].attrs['units'] = 'kilometers'
ds_sweep0['range'].attrs['units'] = 'kilometers'
```

```python
# plot DBZ with range and RHI altitude
fig, ax = plt.subplots(figsize=(20, 4))
ds_sweep0.DBZ.plot.pcolormesh(x='range', y='rhi_alt', cmap="ChaseSpectral", levels=np.arange(-10, 55, 1), ax=ax,add_colorbar=False)
ax.set_xlabel("Range (km)")
ax.set_ylabel("Altitude (km)")
plt.colorbar(ax.collections[0], ax=ax, label="Reflectivity (dBZ)")
ax.set_ylim(0,15)
ax.set_xlim(0,150)
plt.show()
```
